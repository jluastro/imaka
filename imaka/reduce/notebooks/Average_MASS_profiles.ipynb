# Average MASS profiles

```python
from imaka.analysis import plot_stats
from imaka.reduce import massdimm
```

```python
UT_date = '20170109'
massdimm_dir = '/Users/dorafohring/Desktop/imaka/massdimm/'
stats_dir = '/Users/dorafohring/Desktop/imaka/data/'
```

Get stats FITS tables

```python
plot_stats.fetch_stats_from_onaga([UT_date], output_root=stats_dir)
```

Get MASS/DIMM files

```python
save_suffixes = ['open', 'closed'] #'tt','closeda', 'closedb', 'closedc', 'closede']
```

```python
massdimm.fetch_data(utDate=UT_date, saveTo=massdimm_dir)
for suffix in save_suffixes:
    massdimm.append_mass_dimm(stats_dir + '{0}/fli/reduce/stats/stats_{1}.fits'.format(UT_date, suffix), massdimm_dir)
```

Plots:

```python
plot_stats.plot_profile('20170109', suffixes=['open', 'closed'], root_dir=stats_dir)
```
